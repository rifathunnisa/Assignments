```python
import pandas as pd
import numpy as np
zoo=pd.read_csv("Zoo.csv")
zoo
```

```python
del zoo['animal name']
```

```python
zoo
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(zoo,test_size=0.20)

from sklearn.neighbors import KNeighborsClassifier as KNC
neigh=KNC(n_neighbors=5)                                    #5 has been taken as random number
neigh.fit(train.iloc[:,0:15],train.iloc[:,16])
train_acc=np.mean(neigh.predict(train.iloc[:,0:15])==train.iloc[:,16])
test_acc=np.mean(neigh.predict(test.iloc[:,0:15])==test.iloc[:,16])
test_acc
```

```python
neigh=KNC(n_neighbors=3)                                    #3 has been taken as random number
neigh.fit(train.iloc[:,0:15],train.iloc[:,16])
train_acc=np.mean(neigh.predict(train.iloc[:,0:15])==train.iloc[:,16])
test_acc=np.mean(neigh.predict(test.iloc[:,0:15])==test.iloc[:,16])
test_acc
```

```python
acc=[]
for i in range(3,40,2):
    neigh=KNC(n_neighbors=i)
    neigh.fit(train.iloc[:,0:15],train.iloc[:,16])
    train_acc=np.mean(neigh.predict(train.iloc[:,0:15])==train.iloc[:,16])
    test_acct=np.mean(neigh.predict(test.iloc[:,0:15])==test.iloc[:,16])
    acc.append([train_acc,test_acc])
import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(3,40,2),[i[0] for i in acc],"ro-")
```
